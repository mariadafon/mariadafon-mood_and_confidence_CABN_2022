```python
def orderOfMagnitude(number):
    if number>0:
        return m.floor(m.log(number, 10))
    elif number==0:
        return 0
    else:
        return m.floor(m.log(-number, 10))
def roundP(p_value):
    if p_value!=0:
        p = np.round(p_value,-orderOfMagnitude(p_value))
        return p
    else:
        return p_value
```
